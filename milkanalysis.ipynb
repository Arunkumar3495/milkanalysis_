```python
X=df.drop('Grade',axis=1)
y=df['Grade']
```

# Model Building

```python
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=0.3,random_state=0)
```

```python
X_train
```

```python
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
```

```python
X_train_std=scaler.fit_transform(X_train)
X_test_std=scaler.fit_transform(X_test)
```

```python
X_train_std
```

```python
X_train
```

```python
df.columns
```

```python
# import linear regression

from sklearn.linear_model import LogisticRegression
lr = LogisticRegression()
```

```python
lr.fit(X_train,y_train)
```

```python
print("Accuracy score on Training data :",lr.score(X_train,y_train))
print("Accuracy score in test data :", lr.score(X_test,y_test))
```

# confusion Matrix

```python
from sklearn import metrics
```

```python
confusion_matrix = metrics.confusion_matrix
```

```python
y_pred = lr.predict(X_test)
```

```python
cf_matric = confusion_matrix(y_pred,y_test)
```

```python
cf_matric
```

```python
sns.heatmap(cf_matric,annot = True)
```

# classification report

```python
from sklearn.metrics import classification_report
```

```python
print(classification_report(y_test,y_pred))
```

# Decision tree classifier

```python
from sklearn.tree import DecisionTreeClassifier
dtc = DecisionTreeClassifier(criterion='entropy', max_depth = 3,random_state=42)
```

```python
dtc.fit(X_train , y_train)
```

```python
print("Traning Accuracy Score:",dtc.score(X_train,y_train))
print("Testing Accuracy Score:",dtc.score(X_test,y_test))
```

# confusion matrix

```python
y_pred = dtc.predict(X_test)
confusion_matrix(y_pred,y_test)
```

```python
sns.heatmap(confusion_matrix(y_pred,y_test),annot = True)
```

```python
# classification report

print(classification_report(y_pred,y_test))
```

# visualiation tree

```python
from sklearn import tree
tree.plot_tree(dtc.fit(X_train,y_train))
plt.show()
```

# Random Forest tree

```python
from sklearn.ensemble import RandomForestClassifier
rfc= RandomForestClassifier()
rfc.fit(X_train,y_train)
```

```python
print("Training Accuracy Score:",rfc.score(X_train,y_train))
print("Testing Accuracy Score:",rfc.score(X_test,y_test))
```

```python
y_pred = rfc.predict(X_test)
print(y_pred)
```

```python
from sklearn.metrics import accuracy_score
print("accurarcy Score on test data:",accuracy_score(y_pred,y_test))
```

# classification report

# knn algorithm

```python
from sklearn.neighbors import KNeighborsClassifier
knn=KNeighborsClassifier()
knn.fit(X_train,y_train)
```

```python
print("Training Data acuuracy score:",knn.score(X_train,y_train))
print("testing data accuracy score:",knn.score(X_test,y_test))
```

```python
knn_pred = knn.predict(X_test)
print("Accuracy Score on KNN :",accuracy_score(y_test,y_pred))
```
